```python
data = pd.read_csv('train.csv', na_values='?')
data
```

```python
data = data.set_index('PassengerId')
data
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name =='object']
numerical_columns = data.mode(numeric_only=True).columns.values.tolist()
numerical_columns.remove('Survived')
print(categorical_columns)
print(numerical_columns)
```

```python
data[categorical_columns].describe()
```

```python
data.describe(include=[object])
```

```python
for c in categorical_columns:
    print(data[c].unique())
```

```python
data.count(axis=0)
```

```python
data['Embarked'].describe()
```

```python
data['Embarked'] = data['Embarked'].fillna('S')
```

```python
data.drop(columns='Cabin',inplace=True)
```

```python
categorical_columns.pop(3)
```

```python
categorical_columns
```

```python
data_describe = data.describe(include=[object])
binary_columns = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

```python
data['Embarked'].unique()
```

```python
data_nonbinary = pd.get_dummies(data['Embarked'])
data_nonbinary
```

```python
data.drop(columns=['Name','Ticket'],inplace=True)
```

```python
data['Sex'] = data['Sex'].replace(to_replace='male', value=1)
data['Sex'] = data['Sex'].replace(to_replace='female', value=0)
```

```python
data_numerical = data[numerical_columns]
data_numerical = (data_numerical - data_numerical.mean()) /data_numerical.std()
data_numerical.describe()
```

```python
data
```

```python
survived = data['Survived']
data = pd.concat((data_numerical, data[binary_columns], data_nonbinary), axis=1)
data = pd.DataFrame(data, dtype=float)
print(data.shape)
print(data.columns)
```

```python
from sklearn.model_selection import train_test_split
X = data # Выбрасываем столбец 'Survived'.
y = survived
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.67, random_state=42)
```

```python
feature_names = X_train.columns
print(feature_names)

print(X_train.shape)
print(y_train.shape)
N, d = X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
```

```python
y_train_pred=model.predict(X_train)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_train, y_train_pred))
```

```python
y_test_pred=model.predict(X_test)
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_test_pred))
```

```python
import matplotlib.pyplot as plt
importances = model.feature_importances_
feature_names = X_test.columns
forest_importances = pd.Series(importances, index=feature_names)
fig, ax = plt.subplots()
forest_importances.plot.bar(ax=ax)
ax.set_title("Feature importances ")
fig.tight_layout()
```
